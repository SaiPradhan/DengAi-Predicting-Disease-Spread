```python
max_pam_iq=train4_iq.precipitation_amt_mm.quantile(0.999)
max_rpakpm_iq=train4_iq.reanalysis_precip_amt_kg_per_m2.quantile(0.999)
min_satc_iq=train4_iq.station_avg_temp_c.quantile(0.001)
max_smtc_iq=train4_iq.station_max_temp_c.quantile(0.999)
max_som_iq=train4_iq.station_precip_mm.quantile(0.999)
```

Now train4_sj

```python
train4_sj.quantile(0.999)
```

```python
max_ratk_sj=train4_sj.reanalysis_air_temp_k.quantile(0.999)
max_ravtk_sj=train4_sj.reanalysis_avg_temp_k.quantile(0.999)
max_rpakm_sj=train4_sj.reanalysis_precip_amt_kg_per_m2.quantile(0.999)
max_rspam_sj=train4_sj.reanalysis_sat_precip_amt_mm.quantile(0.999)
max_spm_sj=train4_sj.station_precip_mm.quantile(0.999)
```

```python
train5_iq=train4_iq[(train4_iq['precipitation_amt_mm']<max_pam_iq) & (train4_iq['reanalysis_precip_amt_kg_per_m2']<max_rpakpm_iq) & (train4_iq['station_avg_temp_c']>min_satc_iq) & (train4_iq['station_max_temp_c']<max_smtc_iq) & (train4_iq['station_precip_mm']<max_som_iq)]
train5_iq.shape
```

```python
train5_sj=train4_sj[(train4_sj['reanalysis_air_temp_k']<max_ratk_sj) & (train4_sj['reanalysis_avg_temp_k']<max_ravtk_sj) & (train4_sj['reanalysis_precip_amt_kg_per_m2']<max_rpakm_sj) & (train4_sj['reanalysis_sat_precip_amt_mm']<max_rspam_sj) & (train4_sj['station_precip_mm']<max_spm_sj)]
train5_sj.shape
```
